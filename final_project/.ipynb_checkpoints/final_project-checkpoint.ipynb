# How do alliance ratings and implementation of Imagery Exposure during psychotherapy influence primary outcome for patients with GAS?

### Theoretical Background

Generalized anxiety disorder (GAD) is one of the most prevalent anxiety disorders (Asmundson & Asmundson, 2018; Starcevic, Portman & Beck, 2012). The lifetime prevalence of GAD varies between 2% up to 20% and has a high comorbidity with other mental disorders (Ruscio et al., 2017; Angst et al., 2016). Uncontrollable worry as a primary symptom of GAD can be seen as a dysfunctional cognitive strategy to cope with anxiety (Borkovec, Alcaine, & Behar, 2004; Newman, & Llera, 2011; Roemer, Salters, Raffa, & Orsillo, 2005). Moreover, the worrisome thoughts may drain an individuals available cognitive resources (Hirsch & Mathews, 2012; Held, Vîslă, Zinbarg, Wolfer, & Flückiger, 2020).
This dataframe contains information about a clinical outpatient sample of 80 patients, treated by 20 therapists. Patients are all diagnosed with GAD as a main diagnosis. 
The patients are randomized allocated to the therapists and the therapists all implement the updated “Mastery of Your Anxiety and Worry” package (MAW- package; Zinbarg, Craske & Barlow, 2006; Craske & Barlow, 2006), which is widely used for the treatment of GAD and was evaluated in randomized-controlled trials (e.g., Avdagic, Morrissey & Boschen, 2014; Flückiger, et al., 2016; Westra, Constantino & Antony, 2016; Zinbarg, Lee & Yoon, 2007). During the treatment (20 sessions), the most relevant psychotherapy process variable alliance was measured after each session. The influence of the working alliance on outcome is well documented within many studies and is known as a "common factor" of different bona-fide psychotherapies. 
An other interesting topic is the treatment fidelity/adherence. Within every 5 sessions, the therapist adherence was measured. For the analysis, one major component for GAD treatment was selected: Imagery Exposure. 
Beside the general adherence, two different implementation strategies of how therapists explore occurring change within a well-introduced CBT approach for GAD (i.e., the MAW-package) are contrasted into the following two implementation conditions: (a) State Of The Art (SOTA) implementation, where the check-in phase at the beginning of each session primarily focuses on CBT session-agenda setting and review of last session’s homework assignments that may cover a short exploration of changes; (b) Prolonged Focus on Change (PFC) implementation, where the check-in phase allows the therapist a more prolonged focus on exploring occurring changes. The main difference between the two bonafide CBT implementation conditions lies in the attentional focus on and in the systematization of how the therapists explore occurring changes within the usually recommended session-structure (Zinbarg et al., 2006).

measuring instruments

WAI: working alliance inventory
BPSB: bern post session report
PSWQ: penn state worry questionnaire

#### lasso regression

lasso regression to reduce parameters

```r
set.seed(1)
train <- na.omit(final) %>% sample_frac(0.5)
test <- na.omit(final) %>% setdiff(train)
x_train = model.matrix(as.numeric(PSWQ_post) ~., train)[ ,-1]
x_test = model.matrix(as.numeric(PSWQ_post) ~., test)[ ,-1]

y_train = train %>%
  select(PSWQ_post) %>%
  unlist() %>%
  as.numeric()

y_test = test %>%
  select(PSWQ_post) %>%
  unlist() %>%
  as.numeric()
```

```r
library(glmnet)
install.packages("lmerTest")
library(lmerTest)
```

```r
lasso_mod <- glmnet(x_train, 
                   y_train, 
                   alpha = 1)
plot(lasso_mod)

set.seed(1)
cv.out = cv.glmnet(x_train, y_train, alpha = 1)
plot(cv.out)
bestlam = cv.out$lambda.min
lasso_pred = predict(lasso_mod, s = bestlam, newx = x_test)
mean((lasso_pred - y_test)^2)
```

```r
x = model.matrix(PSWQ_post ~., na.omit(final))[,-1]
```

```r
y = na.omit(final) %>%
  select(PSWQ_post) %>%
  unlist() %>%
  as.numeric()
```

```r
out <- glmnet(x, y, alpha = 1) 
lasso_coef <- predict(out, type = "coefficients", s = bestlam)[1:21,]
lasso_coef

lasso_coef[lasso_coef != 0]
```

predictos: patient_ID, therapist_ID, session, gender, comorbidities, condition, PSWQ_pre, BPSB_Bewaeltigung_mean, BPSB_Klaerung_mean, BPSB_Aufgehoben_mean, BPSB_Direktiv_mean

#### multiple linear regression

final model with the predictors from the lasso regression

```r
model.6 <- lm(PSWQ_post ~ PSWQ_pre + BPSB_Bewaeltigung_mean + BPSB_Klaerung_mean + BPSB_Aufgehoben_mean + BPSB_Direktiv_mean + gender + comorbidities + condition, data = final)
summary(model.6)
```

if PSWQ_post increases, PSWQ_pre and BPSB_direktiv increases for 1 unit
or gender = female, condition = B
if PSWQ_post decreases, BPSB_Bewaeltigung and BPSB_Klaerung increase for 1 unit

#### hierarchical model

add therapist level

```r
model.7 <- lmer(PSWQ_post ~ patient_ID + PSWQ_pre + session + BPSB_Bewaeltigung_mean + BPSB_Klaerung_mean + BPSB_Aufgehoben_mean + BPSB_Direktiv_mean + gender + comorbidities + condition + (1|therapist_ID), data = final)
summary(model.7)
```

```r
install.packages("lattice")
```

```r
library(lattice)
```

```r
dotplot(ranef(model.7, condVar = TRUE))
```

illustration of the therapists effect:
therapist-specific intercepts, e.g. patients with-in therapist 110 show the biggest symptom reduction and patients with-in therapist 113 the smallest. 

```r
performance::icc(model.7)
```

The "adjusted" ICC only considers the random effects, while the "conditional" ICC also takes the fixed effects variances into account.

the random effect for therapists explaine 34.4% of the variance, taking the fixed effects into account the random intercept accounts for 21.9% of the variance in the outcome.

```r
BIC(model.6, model.7)
```

the model with the therapist level better fit than the model without the therapist level

#### Bootstrapping

estimate the population mean of the PSWQ_post score and standard error

```r
mu <- mean(na.omit(final$PSWQ_post))
mu
```

```r
se.mu <- sd(na.omit(final$PSWQ_post))/sqrt(dim(final)[1])
se.mu
```

estimate the sd with bootstrap

```r
library(boot)
set.seed(1)
boot.fn <- function(data, index)return(mean(data[index])) 
boot(na.omit(final$PSWQ_post), boot.fn, 1000)
```

the standard error from the bootstrap is a bit higher than the one from the prior calculation

95 % confidence interval for the mean of PSWQ_post, based on bootstrap estimate

```r
c(48.5 - 2 * 0.57, 48.5 + 2 * 0.57)
```

```r
t.test(na.omit(final$PSWQ_post))
```

the confidence interval is the same for the bootstrap and the t-test

comparing the median

```r
median(na.omit(final$PSWQ_post))
```

```r
set.seed(1)
boot.fn <- function(data, index) return(median(data[index])) 
boot(na.omit(final$PSWQ_post), boot.fn, 1000)
```

its exactly the same
